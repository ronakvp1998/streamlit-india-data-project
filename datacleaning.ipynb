```python
census = pd.read_csv('india-districts-census-2011.csv')
census.head()
```

```python
census.columns
```

```python
cols = ['District code','District name','Population', 'Male','Female', 'Literate','Households_with_Internet']
```

```python
census = census[cols]
```

```python
# merge on district

final_df = latlong.merge(census,left_on='District',right_on='District name').drop(columns=['District name'])
```

```python
final_df['sex_ratio'] = (final_df['Female']/final_df['Male'])*100
```

```python
final_df['Literacy_rate'] = final_df['Literate']/final_df['Population']
```

```python
final_df.head()
```

```python
final_df.drop(columns = ['Litracy_rate','Male','Female','Literate'],inplace=True)
```

```python
final_df.to_csv('india.csv',index='False')
```
